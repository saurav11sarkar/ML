```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_x = LabelEncoder()
```

```python
x['workclass'] = le_x.fit_transform(x.workclass)
x['education'] = le_x.fit_transform(x.education)
x['marital_status'] = le_x.fit_transform(x.marital_status)
x['occupation'] = le_x.fit_transform(x.occupation)
x['relationship'] = le_x.fit_transform(x.relationship)
x['race'] = le_x.fit_transform(x.race)
x['sex'] = le_x.fit_transform(x.sex)
x['native_country'] = le_x.fit_transform(x.native_country)
```
